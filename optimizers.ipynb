```python
import numpy as np
import matplotlib.pyplot as plt
# Define the function and its gradient
def f(x):
    return x**2  # Function

def compute_gradient(f, x):
    h = 1e-6  # Small step for numerical approximation
    return (f(x + h) - f(x)) / h  # Numerical derivative (finite difference method)

# Initialize parameters
x = 10  # Start at x = 10
learning_rate = 0.1  # Learning rate
momentum = 0.9  # Momentum coefficient
velocity = 0  # Initialize velocity
iterations = 10
# Store x values and f(x) values for plotting
x_values = [x]
f_values = [f(x)]

# Gradient descent loop
for i in range(iterations):
    gradient = compute_gradient(f, x)  # Compute gradient dynamically
    x = x - learning_rate * gradient  # Update x
    x_values.append(x)  # Record x
    f_values.append(f(x))  # Record f(x)
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
```

```python
# Plotting the function and gradient descent path
x_range = np.linspace(-11, 11, 500)  # Range of x values for plotting
f_range = f(x_range)  # Corresponding f(x) values

plt.figure(figsize=(10, 6))

# Plot the function
plt.plot(x_range, f_range, label="$f(x) = x^2$", color="blue", linewidth=2)

# Plot the gradient descent steps
plt.scatter(x_values, f_values, color="red", label="Gradient Descent Steps", zorder=5)
plt.plot(x_values, f_values, linestyle="--", color="red", alpha=0.7)

# Mark the starting point
plt.scatter([x_values[0]], [f_values[0]], color="green", label="Start", s=100, zorder=6)

# Mark the ending point
plt.scatter([x_values[-1]], [f_values[-1]], color="purple", label="End", s=100, zorder=6)

# Labels and legend
plt.title("Gradient Descent Visualization", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)
plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
plt.axvline(0, color="black", linewidth=0.7, linestyle="--")
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

```python
def plot_optimization_path(x_values, f_values):
    # Plot the function f(x) = x^2
    x_vals = np.linspace(-12, 12, 400)  # Range of x values for plotting
    y_vals = x_vals**2  # Corresponding f(x) values

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=r'$f(x) = x^2$', color="blue")  # Plot the function f(x)
    plt.scatter(x_values, f_values, color='red', label='Momentum Path', zorder=5)  # Plot the path of gradient descent with momentum
    plt.plot(x_values, f_values, color='red', linestyle='dotted', label='Path of Descent')  # Connect the points to visualize the descent path

    # Add labels, title, and grid
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
x = 10  # Start at x = 10
learning_rate = 0.1  # Learning rate
m = 0 
iterations = 10
beta = 0.999
epsilon=1e-8
# Store x values and f(x) values for plotting
x_values = [x]
f_values = [f(x)]

for i in range(1, iterations + 1):
    gradient = compute_gradient(f, x)
        
    # Update rms and parameter
    m = beta * m + (1 - beta) * gradient**2
    x -= (learning_rate / (np.sqrt(m) + epsilon)) * gradient
        
    # Store values for plotting
    x_values.append(x)
    f_values.append(f(x))
        
    print(f"Iteration {i}: x = {x:.5f}, f(x) = {f(x):.5f}, m = {m:.5f}")
```

```python
x = 10  # Start at x = 10
learning_rate = 0.1  # Learning rate
m = 0  # Momentum coefficient
v = 0  # Initialize velocity
iterations = 250
beta1=0.9
beta2=0.99999999
epsilon=1e-8
# Store x values and f(x) values for plotting
x_values = [x]
f_values = [f(x)]
```

```python
# To store the history for plotting
# Training loop for Adam optimizer
for t in range(1,iterations+1):
    gradient = compute_gradient(f,x)
        
    # Update biased first moment estimate
    m = beta1 * m + (1 - beta1) * gradient
        
    # Update biased second moment estimate
    v = beta2 * v + (1 - beta2) * (gradient**2)
        # Compute bias-corrected first and second moments
    m_hat = m / (1 - beta1**t)
    v_hat = v / (1 - beta2**t)
        
        # Update parameters
    x -= learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
    x_values.append(x)
    f_values.append(x**2)
    print(f"Iteration {t}: x = {x}, f(x) = {f(x)}, velocity = {v}, momentum = {m}")



```

```python
def plot_adam_optimizer(x_values, f_values):
    x_range = np.linspace(-11, 11, 500)  # Range for plotting the function
    f_range = f(x_range)  # Corresponding f(x) values

    plt.figure(figsize=(12, 8))

    # Plot the function f(x) = x^2
    plt.plot(x_range, f_range, label="$f(x) = x^2$", color="blue", linewidth=2)

    # Plot the Adam optimization steps
    plt.scatter(x_values, f_values, color="red", label="Adam Optimization Steps", zorder=5)
    plt.plot(x_values, f_values, linestyle="--", color="red", alpha=0.7)

    # Mark the starting point
    plt.scatter([x_values[0]], [f_values[0]], color="green", label="Start", s=100, zorder=6)

    # Mark the ending point
    plt.scatter([x_values[-1]], [f_values[-1]], color="purple", label="End", s=100, zorder=6)

    # Labels and legend
    plt.title("Adam Optimizer Visualization", fontsize=16)
    plt.xlabel("x", fontsize=14)
    plt.ylabel("f(x)", fontsize=14)
    plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.7, linestyle="--")
    plt.legend(fontsize=12)
    plt.grid(alpha=0.3)

    # Show the plot
    plt.show()
```

```python
plot_adam_optimizer(x_values, f_values)
```
